## SQLite Database

First let's store the data in a simple SQLite database.



How about a helper class to interface with the database:

```python
import os
import sqlite3

from pandas import DataFrame, read_sql_query

DB_FILEPATH = "my_database.db" # for example

class BaseDatabase:
    """A base interface into SQLite database."""

    def __init__(self, filepath=DB_FILEPATH):
        """Params:
            filepath (str) : path to the database that will be created
        """
        self.filepath = filepath
        print("------------------")
        print("DB FILEPATH:", os.path.abspath(self.filepath))

        self.connection = sqlite3.connect(self.filepath)
        self.connection.row_factory = sqlite3.Row
        print("CONNECTION:", self.connection)

        self.cursor = self.connection.cursor()
        print("CURSOR:", self.cursor)

    def insert_df(self, df:DataFrame, table_name:str): # auto_increment=False
        """Inserts data into a given table.
            If table does not exist, will create the table.
            Normally with a SQL-only solution we would need to create the table first, with a given schema.
            But pandas.DataFrame.to_sql infers the schema from the dataframe schema.

            Params:

                table_name (str) : name of table to insert data into

                df (DataFrame) : the data to save or append in the given table
        """
        # https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_sql.html
        df.to_sql(table_name, con=self.connection,
            if_exists="append", # append to existing tables (don't throw error)
            index=False
        )

    def query_to_df(self, sql:str) -> DataFrame:
        """Executes a query and returns the results as a dataframe"""
        return read_sql_query(sql, con=self.connection)


db = BaseDatabase()
```

It is possible to insert data without specifying a table schema:

```python
# no schema is possible, but lacks proper primary keys (if one likes that kind of thing)
#db.insert_df(df=books_df, table_name="other")
#db.insert_df(df=more_books_df, table_name="other")
#db.query_to_df("SELECT * FROM other;")
```

... but ...

Specifying a schema structure will implement the right primary key structure with an auto-incremented integer identifier, even if we insert data in multiple batches. And we can take advantage of other database conventions like timestamp columns.

Create the table:

```python
# set up a more well defined schema:
db.cursor.execute("DROP TABLE IF EXISTS books;")
db.cursor.execute("""
    CREATE TABLE IF NOT EXISTS books (

        id              INTEGER PRIMARY KEY,

        title           VARCHAR(255),
        author          VARCHAR(255),
        year_published  INTEGER,
        publisher       VARCHAR(255),
        genre           VARCHAR(255),
        isbn            VARCHAR(255),
        language        VARCHAR(255),
        page_count      INTEGER,

        created_at      DATETIME NOT NULL DEFAULT (strftime('%Y-%m-%d %H:%M:%f', 'now'))
    );
""")
db.connection.commit() # saves the query results (use when writing data to the database)
```

Here is a helper class for interfacing with a Google BigQuery Database. Update the `PROJECT_NAME` variable with your own project name:

```python
# https://github.com/s2t2/tweet-analysis-2023/blob/main/app/bq_service.py

from datetime import datetime
from google.cloud import bigquery
from pandas import DataFrame

PROJECT_NAME = "tweet-research-public" # use your own project name!


class BigQueryService():
    def __init__(self, project_name=PROJECT_NAME):
        self.client = bigquery.Client(project=project_name)

    def execute_query(self, sql, verbose=True):
        if verbose == True:
            print(sql)
        job = self.client.query(sql)
        return job.result()

    def query_to_df(self, sql, verbose=True):
        """high-level wrapper to return a DataFrame"""
        results = self.execute_query(sql, verbose=verbose)
        records = [dict(row) for row in list(results)]
        df = DataFrame(records)
        return df

    @staticmethod
    def split_into_batches(my_list, batch_size=10_000):
        """Splits a list into evenly sized batches"""
        # h/t: https://stackoverflow.com/questions/312443/how-do-you-split-a-list-into-evenly-sized-chunks
        for i in range(0, len(my_list), batch_size):
            yield my_list[i : i + batch_size]

    @staticmethod
    def generate_timestamp(dt=None):
        """Formats datetime object for storing in BigQuery. Uses current time by default. """
        dt = dt or datetime.now()
        return dt.strftime("%Y-%m-%d %H:%M:%S")

    def insert_records_in_batches(self, table, records):
        """
        Inserts records in batches because attempting to insert too many rows at once
            may result in google.api_core.exceptions.BadRequest: 400

        Params:
            table (table ID string, Table, or TableReference) like project-name
            records (list of dictionaries)
        """
        rows_to_insert = [list(d.values()) for d in records]
        #errors = self.client.insert_rows(table, rows_to_insert)
        errors = []
        batches = list(self.split_into_batches(rows_to_insert, batch_size=5_000))
        for batch in batches:
            errors += self.client.insert_rows(table, batch)
        return errors


bq = BigQueryService()
print(bq)
```

Create table on bigquery (syntax is a little different from one
 type of database to another).

```python
DATABASE_NAME = "example_database" # or your own database you created
DATASET_ADDRESS = f"{PROJECT_NAME}.{DATABASE_NAME}"

sql = f"""
    CREATE TABLE IF NOT EXISTS `{DATASET_ADDRESS}.books` (

        -- there are no primary keys in bigquery! it is write-oriented

        title STRING,
        author STRING,
        year_published INTEGER,
        publisher STRING,
        genre STRING,
        isbn STRING,
        language STRING,
        page_count INTEGER,

        -- no auto-timestamps either. although you can use datetime columns in general

    );
"""
bq.execute_query(sql)
```

```python
# API call reference to the table that got created (need when inserting records later)
books_table = bq.client.get_table(f"{DATASET_ADDRESS}.books")
print(type(books_table))
```

```python
errors = bq.insert_records_in_batches(records=books, table=books_table)
print(errors)
```
